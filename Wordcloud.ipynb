```python
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
import numpy as np
import re
```

```python
text = open('example.txt')
with open('example.txt', encoding='cp1252') as f:
    text =''.join(f.readlines())
    
text = re.sub(r'==.*?==+','',text)
text = text.replace('\n','')
# print(text)
```

```python
# function to plot cloud
def plot_cloud(wc):
    plt.figure(figsize=(30,40))
    plt.imshow(wc)
    plt.axis("off")
```

```python
wc = WordCloud(width=3000, height=2000, max_words = 50, random_state=42, background_color='white', colormap='coolwarm', 
               collocations=False, stopwords=swords)
```

```python
wc.generate(text)
plot_cloud(wc)
```

```python
wc.to_file('gs_strategy_nomask.png')
```
